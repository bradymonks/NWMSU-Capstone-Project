```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('AthleteDatabase.csv')
```

```python
birthplace_league = pd.crosstab(df['BirthPlace'], df['League'], normalize='index') * 100
print(birthplace_league)
```

```python
plt.figure(figsize=(12, 8))
sns.heatmap(birthplace_league, annot=True, cmap="YlGnBu")
plt.title('Percentage of Players in Each League by BirthPlace')
plt.xlabel('League')
plt.ylabel('BirthPlace')
plt.show()
```

```python
college_league = pd.crosstab(df['College'], df['League'], normalize='index') * 100
print(college_league)
```

```python
plt.figure(figsize=(12, 8))
sns.heatmap(college_league, annot=True, cmap="YlGnBu")
plt.title('Percentage of Players in Each League by College')
plt.xlabel('League')
plt.ylabel('College')
plt.show()
```

```python
college_rankings = pd.read_csv('FinalCollegeRankings.csv')
birthplace_rankings = pd.read_csv('FinalBirthPlaceRankings.csv')
```

```python
df_with_college_rankings = pd.merge(df, college_rankings, on='College', how='left')
```

```python
df_with_rankings = pd.merge(df_with_college_rankings, birthplace_rankings, on='BirthPlace', how='left')
```

```python
import numpy as np

plt.figure(figsize=(12, 8))
sns.scatterplot(data=df_with_rankings, x='MLB_x', y='MLB_y', alpha=0.5)
plt.title('College vs. BirthPlace Rankings in MLB')
plt.xlabel('College MLB Ranking')
plt.ylabel('BirthPlace MLB Ranking')
plt.plot(np.unique(df_with_rankings['MLB_x']), np.poly1d(np.polyfit(df_with_rankings['MLB_x'], df_with_rankings['MLB_y'], 1))(np.unique(df_with_rankings['MLB_x'])), color='r')
plt.show()
```
